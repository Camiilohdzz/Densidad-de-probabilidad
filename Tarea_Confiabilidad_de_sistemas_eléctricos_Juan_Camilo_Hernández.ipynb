# **Punto A**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.optimize import minimize
from scipy import stats 
import seaborn as sns 
```

```python
mu, sigma = 1.5, 0.5 # media y desvio estandar
datos = np.random.normal(mu, sigma, 1000) #creando muestra de datos


FDP_normal = stats.norm(0.5, 4).pdf(datos) # FDP
plt.plot(datos, FDP_normal, label='FDP nomal')
plt.title('Funci√≥n de Densidad de Probabilidad')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
mu, sigma = 1.5, 0.5 # media y desvio estandar
datos = np.random.normal(mu, sigma, 1000) #creando muestra de datos


FDP_normal = stats.norm(0, 30).pdf(datos) # FDP
plt.plot(datos, FDP_normal, label='FDP nomal')
plt.title('Funci√≥n de Densidad de Probabilidad')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```
